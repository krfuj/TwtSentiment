# Data Cleaning and Preprocessing
The code is going to calculate the upper bound for identifying outliers in a dataset.The code `duplicate_rows = df[df.duplicated()]` will identify and store the duplicate rows in the DataFrame `df`. The `df.duplicated()` function will return a boolean Series indicating whether each row is a duplicate or not. By passing this boolean Series as a filter to the DataFrame `df`, only the duplicate rows are selected and stored in the variable `duplicate_rows`.

The code will calculating the number of missing values in each column of the DataFrame `df`. It uses the `isna()` method to check for missing values and the `sum()` method to calculate the total number of missing values in each column. The result will store in the variable `missing_values`.

Later we will remove the specific outlier value from the dataframe `df`. It filters out rows where the value in the "tweet_id" column is equal to a specific value.

```python
# Identify outliers
outliers = df.describe().loc["75%"] + 1.5 * (df.describe().loc["75%"] - df.describe().loc["25%"])
outliers
```

```python
# Identify inconsistencies
duplicate_rows = df[df.duplicated()]
duplicate_rows
```

```python
# Identify missing values
missing_values = df.isna().sum()
missing_values
```

```python
# Remove the outlier value
df = df[df["tweet_id"] != 19210.0]

# removing duplicates
df.drop_duplicates(inplace = True)

# Remove the duplicate rows
df = df.drop_duplicates()

#This will remove any character in the â€˜tweetsâ€™ column that is not a letter (upper or lower case), a number, a space, or a punctuation character (.,;?!$%^&*()-_+=[]{}'|"<>`~) with an empty string
df['tweets'] = df['tweets'].astype(str)
df['tweets'] = df['tweets'].str.replace('[^a-zA-Z0-9\s.,;?!$%^&*()-_+=[]{}\'"|<>`~]', '', regex=True)


#fill 0 in mum
df['tweets'] = df['tweets'].fillna(0)

# Save the dataframe
df.to_csv("tweet_clean.csv", index=False)
```

# Exploratory data analysis (EDA)

`df.describe().T` will transposing the output of the `describe()` method on a DataFrame `df`. The `describe()` method provides summary statistics of the numerical columns in the DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles. By applying `.T` after `describe()`, the output is transposed, meaning the rows become columns and vice versa. This can be useful for better readability or for further analysis of the summary statistics.
df.describe().T

The code will be using the `OrdinalEncoder` class from the `sklearn.preprocessing` module to convert categorical values in the specified columns of a DataFrame (`df`) into ordinal integers.

```python
df.describe().T
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()

# Every single column with categorical values will be converted.
object_cols = ['tweet_id', 'sentiments', 'tweets']
df[object_cols] = df[object_cols].astype(str)

df[object_cols] = ordinal_encoder.fit_transform(df[object_cols])

df.head()
```
